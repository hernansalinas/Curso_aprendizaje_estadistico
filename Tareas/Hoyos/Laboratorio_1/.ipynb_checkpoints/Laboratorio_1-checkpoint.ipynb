# 1. Importación de los datos

```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
breastdf=pd.read_csv('Breast_Cancer.csv')
breastdf #Datos originales
```

# 2. Renombramiento de las columnas

```python
column=[col.replace("_", " ") for col in breastdf.columns] #Lista nombres en el df sin guiones bajos
column_splitted=[c.split() for c in column] 
column_capitalized=[[a.lower().capitalize() for a in b] for b in column_splitted]#Se hacen mayúsuclas las iniciales de cada palabra
column_joined=["".join(a) for a in column_capitalized] #Se suprimen los espacios entre palabras
breastdf.columns=column_joined #Se modifican los nombres de las columnas en el df
```

```python
del(breastdf["Unnamed:32"])  #Se suprime una columna con oslo entradas nan
```

# 8. Creación Diagnosis Numeric

```python
Diagnosis=breastdf.Diagnosis
DiagnosisNumeric=[0  if i=='B'  else 1 for i in breastdf.Diagnosis]
breastdf['DiagnosisNumeric']=DiagnosisNumeric
```

# 9. Normalización de cada columna

```python
del(breastdf["Diagnosis"])
```

```python
for i in breastdf.columns:
    for j in range(len(breastdf.RadiusMean)):
        breastdf.loc[:,i][j]= (breastdf.loc[:,i][j]- np.mean(breastdf.loc[:,i]))/np.std(breastdf.loc[:,i])  
```

breastdf.loc[:,i][j] toma la i-ésima columna del dataframe, por ejemplo DiagnosisNumeric y de ella toma la entrada j-ésima. El ciclo anidado for entonces asocia a cada valor de las columnas un valor sustituto que se obtiene de la razón entre  la resta de cada valor con la media de esa columna y la desviación estándar de ella.

```python
breastdf["Diagnosis"]=Diagnosis #Se recupera la columna Diagnosis que se eliminó para realizar la normalización
```

# Histograma para Radius Mean con las etiquetas B y M

```python
color={'B':'orange', 'M':'blue'}

plt.figure(figsize=(15,10))
for name, group in breastdf.groupby('Diagnosis'):
    plt.hist(group['RadiusMean'], label=name, color=color[name])

plt.tick_params(labelsize="15")
plt.legend(loc="upper right", fontsize="18")
plt.ylabel("Frecuencia", fontsize="25")
plt.xlabel("Radius Mean", fontsize="25")
plt.grid()
```

# Diagramas violin

```python
plt.figure(figsize=(30,20))
data = pd.melt(breastdf.iloc[:, 0:9], id_vars="Diagnosis",var_name="Features",value_name="Values")
sns.violinplot(x="Features",y="Values",  hue="Diagnosis",data=data, split=True, inner="quart")
plt.tick_params(labelsize="25")
plt.legend(loc="upper right", fontsize="30")
plt.xticks(rotation=45)
plt.ylabel("Valores",fontsize="25")
plt.xlabel("Características", fontsize="25")
plt.ylim(-3,5) #Para mejor apreciación. si se suprime esta línea se graficas completos
```

# 10. Outliers

```python
#Gráfico boxplot
breastdf.boxplot(column="RadiusMean", by='Diagnosis', sym = 'k.', figsize=(25,12))

plt.xlabel("Diagnosis", fontsize="25")
plt.tick_params(labelsize="25")
```

```python
q1=breastdf.groupby('Diagnosis').quantile(0.25).RadiusMean 
q3=breastdf.groupby('Diagnosis').quantile(0.75).RadiusMean
IQR=q3-q1
```

```python
breastdf_NoOutliers=breastdf[((breastdf['Diagnosis']=='B')&((breastdf['RadiusMean']>(q1.B-1.5*IQR.B))&(breastdf['RadiusMean']<(q3.B+1.5*IQR.B))))|((breastdf['Diagnosis']=='M')&
                                                                                                                      ((breastdf['RadiusMean']>(q1.M-1.5*IQR.M))&(breastdf['RadiusMean']<(q3.M+1.5*IQR.M))))]
```

# 11. Matriz de correlación

```python
plt.figure(figsize=(30,20))
sns.heatmap(breastdf_NoOutliers.corr(), cmap="YlGnBu", annot=True)
plt.tick_params(labelsize="20")
```

# 12. Otros gráficos posibles

Además de los ya usados también pueden construirse diagramas tipo KDE o "kernel density estimation" que se obtiene de la suma de las distribuciones asociadas a cada dato. Estas podrían ser de diversas índoles: cosenoidales, gausianas , triangulares, etc. La distribución obtenida de esta gráfica permite observar una distribución continua a partir de un conjunto discreto de datos. De hecho, estos gráficos se usan como insumo para los diagramas de violin.

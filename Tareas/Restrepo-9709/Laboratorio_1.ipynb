## 1.

```python
temptx = np.loadtxt('Land_and_Ocean_complete.csv', comments = '%')
temptx
```

## 2.

```python
columns = ["Year","Month","MonthlyA","MonthlyUnc","AnnuealA","AnnuealUnc",
           "FiveA","FiveUnc","TenA","TenUnc","TwentyA","TwentyUnc"]
```

```python
temp = {columns[0]:temptx[:,0], columns[1]:temptx[:,1], columns[2]:temptx[:,2], columns[3]:temptx[:,3], 
        columns[4]:temptx[:,4], columns[5]:temptx[:,5], columns[6]:temptx[:,6], columns[7]:temptx[:,7], 
       columns[8]:temptx[:,8], columns[9]:temptx[:,9], columns[10]:temptx[:,10], columns[11]:temptx[:,11]}
```

```python
dftemp = pd.DataFrame(temp)
dftemp
```

## 3.

```python
dftemp.loc[:,'Year'] = [int(dftemp.loc[i,'Year']) for i in range(0,4128)]
dftemp.loc[:,'Month'] = [int(dftemp.loc[i,'Month']) for i in range(0,4128)]
#dftemp
```

```python
dftemp['Date'] =  [str(dftemp.loc[i,'Year']) + str(dftemp.loc[i,'Month'])  for i in range(0,4128)]
#dftemp
```

## 4.

```python
dftemp["Date"] = pd.to_datetime(dftemp["Date"],format='%Y%m')
dftemp
```

## 5.

```python
T_serie = dftemp.groupby(pd.Grouper(key='Date', freq='Y')).mean()
T_serie
```

## 6.

```python
fig = plt.figure(figsize = (8,5))
T_serie.MonthlyA.plot()
plt.ylabel('Anomaly ')
plt.show()
```

## 7.


```python
fig = plt.figure(figsize = (8,5))
T1 = T_serie.loc['1850-01-01':'1920-01-01']
m, b = np.polyfit(T1['Year'], T1['MonthlyA'], 1)

plt.plot(T1['Year'], T1['MonthlyA'])
plt.ylabel('Anomaly ')
plt.plot(T1['Year'], m*T1['Year'] + b, color = 'red')
print('Best fit: m = ',m,', b = ',b )
plt.title('Años 1850-1920')
plt.show()
```

```python
fig = plt.figure(figsize = (8,5))
T1 = T_serie.loc['1920-01-01':'1970-01-01']
m, b = np.polyfit(T1['Year'], T1['MonthlyA'], 1)

plt.plot(T1['Year'], T1['MonthlyA'])
plt.ylabel('Anomaly ')
plt.plot(T1['Year'], m*T1['Year'] + b, color = 'red')
print('Best fit: m = ',m,', b = ',b )
plt.title('Años 1920-1970')
plt.show()
```

```python
fig = plt.figure(figsize = (8,5))
T1 = T_serie.loc['1970-01-01':'2022-01-01']
m, b = np.polyfit(T1['Year'], T1['MonthlyA'], 1)

plt.plot(T1['Year'], T1['MonthlyA'])
plt.ylabel('Anomaly ')
plt.plot(T1['Year'], m*T1['Year'] + b, color = 'red')
print('Best fit: m = ',m,', b = ',b )
plt.title('Años 1970-2022')
plt.show()

print('\n\n\n Para el año 2100 se predice una anomalía de ', 0.018559070761831387*2100-36.59602073728582, '°C')
```


Análisis: Se observa de la gráfica de anomalía respecto al tiempo un ascenso de la temperatura entre 1920-1970 con una pendiente de 0.0044°C/Año y entre 1920-1970 de 0.01856°C/Año. Esto coincide con el aumento de emisiones de combustibles fósiles por parte de la industria mundial como se señala en la sigiente gráfica.


Esta relación aporta evidencia a la influencia de las actividades humanas en el cambio climático. El ajuste lineal pronostica una anomalía de 2.3780°C para el año 2100. Según https://scied.ucar.edu/learning-zone/climate-change-impacts/predictions-future-global-climate se predice una anomalía de aproximadamente 4°C si las actividades industriales continúan sin regulación causando el aumento en el nivel y acidificacón del mar implicando amenazas para los ecosistemas terrestres y marinos.

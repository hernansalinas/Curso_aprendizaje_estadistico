```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
sns.set()
```

##1.
Leer el archivo de datos

```python
url = "http://berkeleyearth.lbl.gov/auto/Global/Land_and_Ocean_complete.txt"
```

```python
data = np.loadtxt(url,comments ="%")
```

##2.
Construir un dataframe con las columnas:
columns = ["Year","Month","MonthlyA","MonthlyUnc","AnnuealA","AnnuealUnc",\
           "FiveA","AnnuealUnc","TenA","AnnuealUnc","TwentyA","AnnuealUnc"]

```python
df = pd.DataFrame(data,columns = ["Year","Month","MonthlyA","MonthlyUnc","AnnuealA","AnnuealUnc","FiveA","AnnuealUnc","TenA","AnnuealUnc","TwentyA","AnnuealUnc"])
#columns = ["Year", "Month",  "Monthly_Anomaly", "Monthly_Unc.",   "Annual_Anomaly", "Annual_Unc.",   "Five-year_Anomaly", "Five-year_Unc.",   "Ten-year_Anomaly", "Ten-year_Unc.",   "Twenty-year_Anomaly", "Twenty-year_Unc."])
```

```python
df
```

##3. 
Construir una nueva columna tipo string llamada Date con la informacion del Year y Month

```python
c = []
d= "_"
for i in range(len(df["Year"])):
  #for j in range(len(df["Month"])):
  a = str(int(df["Year"].iloc[i]))
  b = str(int(df["Month"].iloc[i]))
  c.append(a+b)
```

```python
df=df.assign(Date=c)
df
```

##6.

Realice un gráfico del tiempo como función de MonthlyA.

```python
plt.figure(dpi=100)
plt.plot(df["MonthlyA"],df["Date"],".",ms=0.8)
plt.xlabel(r"$MonthlyA$")
plt.ylabel(r"$Date$")
plt.title("Gráfico del tiempo como función de MonthlyA")
plt.legend()
plt.grid()
```

##7.

¿Que puede concluir?

Si sacamos una gráfica del promedio de la temperatura a lo largo del año por cada año, obtenemos

```python
plt.figure(dpi=100)
plt.plot(np.array(df.groupby(pd.Grouper(key='Date', freq='Y')).MonthlyA.mean()),np.array(df.groupby(pd.Grouper(key='Date', freq='Y')).Year.mean()),"o",ms=0.8)
plt.xlabel(r"$MonthlyA$")
plt.ylabel(r"$Date$")
plt.title("Gráfico del tiempo como función de MonthlyA")
plt.legend()
plt.grid()
```

Como podemos observar en ambas gráficas a lo largo del tiempo la temperatura global de la tierra ha ido aumentando, notandose un aumento mucho más acelerado a partir de los años 80's pareciendose mucho la distribución a una logarítimica. Esto muestra claramente que el calentamiento global ha sido directamente proporcional al aumento de tecnologías, producción global y tasa de natalidad en la tierra, donde aproximadamente en estos años es donde comienza a acelerar mucho más hasta el día de hoy.
